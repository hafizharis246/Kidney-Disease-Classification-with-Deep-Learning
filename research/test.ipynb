## Downloading data from Google Drive

```python
import gdown
```

```python
url = "https://drive.google.com/file/d/1PIrR-mWs00bwZr8aO1vDw0rwISF4yVxP/view?usp=sharing"
```

```python
file_id = url.split("/")[-2]
file_id
```

```python
prefix = 'https://drive.google.com/uc?/export=download&id='
gdown.download(prefix+file_id, "kidney-CT-Scan-data.zip")
```
